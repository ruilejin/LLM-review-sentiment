```python
movie_category=movie_data.groupby(['genres']).count()
movie_category
```

```python
movie_category.sort_values(by=['rotten_tomatoes_link'],ascending=False)
```

```python
categories=movie_category.index
categories
```

Analyse categories.

```python
# Tokenize categories combinations
cate_comb=[]
for i in categories:
  c=i.split(',')
  s=[]
  for b in c:
    b=b.lstrip() # Delete the space on the left.
    s.append(b)
  cate_comb.append(s)
cate_comb
```

```python
#Get a category list where there is no duplication.
category_list=[]
for i in cate_comb:
  for j in i:
    if j in category_list:
      pass
    else:
      category_list.append(j)
```

```python
category_list.sort()
category_list
```

```python
len(category_list)
```

```python
df4=pd.DataFrame(data={'category':cate_comb,'quantity_of_movie':movie_category['rotten_tomatoes_link']})
df4=df4.sort_values(by=['quantity_of_movie'],ascending=False)
```

```python
df4=df4.reset_index(drop=True)
df4
```

Find single category.

```python
single_cate=[]
quantity=[]
for i,j in zip(df4['category'],df4['quantity_of_movie']):
  if len(i)==1:
    print(i,j)
    single_cate.append(i)
    quantity.append(j)
single_cate=[token for i in single_cate for token in i]
```

```python
df_single_cate=pd.DataFrame(data={'genre':single_cate[:10],'quantity':quantity[:10]})
df_single_cate
```

Check the balance.

```python
from collections import Counter
Counter(review_category['review_type'])
```

Sample to smaller shape and adjust balance.

```python
Drama=review_category.loc[review_category['genres'].isin(['Drama'])]
Drama_0=Drama.loc[Drama['review_type'].isin(['Rotten'])]
Drama_1=Drama.loc[Drama['review_type'].isin(['Fresh'])]
Drama_0_samp=Drama_0.sample(n=10000)
Drama_1_samp=Drama_1.sample(n=10000)
Drama1=pd.concat([Drama_0_samp,Drama_1_samp])
Counter(Drama1['review_type'])
```

```python
Comedy=review_category.loc[review_category['genres'].isin(['Comedy'])]
Comedy1=Comedy.sample(n=20000)
Counter(Comedy1['review_type'])
```

```python
Documentary=review_category.loc[review_category['genres'].isin(['Documentary'])]
Documentary_0=Documentary.loc[Documentary['review_type'].isin(['Rotten'])]
Documentary_1=Documentary.loc[Documentary['review_type'].isin(['Fresh'])]
Documentary_1_samp=Documentary_1.sample(n=3000)
Documentary1=pd.concat([Documentary_0,Documentary_1_samp])
```

```python
Counter(Documentary1['review_type'])
```

```python
Horror=review_category.loc[review_category['genres'].isin(['Horror'])]
Horror_0=Horror.loc[Horror['review_type'].isin(['Rotten'])]
Horror_1=Horror.loc[Horror['review_type'].isin(['Fresh'])]
Horror_0_samp=Horror_0.sample(n=9488)
Horror_1_samp=Horror_1.sample(n=9488)
Horror1=pd.concat([Horror_0_samp,Horror_1_samp])
Counter(Horror1['review_type'])
```

```python
data=pd.concat([review_category2,Drama1,Comedy1,Documentary1,Horror1])
```
